```python
import random
import math
import json

# Generate some test data

vocab = ["foo", "bar", "baz", "qux", "quux", "corge", "grault", "garply", "waldo", "fred", "plugh", "xyzzy", "thud", "onetwoonetwo"]
next_tokens = {
    tok: {random.choice(vocab): random.random() for _ in range(10)}
    for tok in vocab
}
# normalize
for tok in vocab:
    total = sum(next_tokens[tok].values())
    for next_tok in next_tokens[tok]:
        next_tokens[tok][next_tok] /= total

jsonl_lines = []
for i in range(50):
    for j in range(random.randint(5, 10)):
        line_tokens = [random.choice(vocab) for _ in range(random.randint(5, 35))]
        jsonl_lines.append({
            "tokens": line_tokens,
            "metrics": {
                "token_length": [len(t) for t in line_tokens],
                "increasing_var": [random.normalvariate() * n for n, _ in enumerate(line_tokens)],
                "increase_with_noise": [random.normalvariate() + n for n, _ in enumerate(line_tokens)],
                "sparse_cos": [random.normalvariate() + math.cos(k * math.pi / 10) if random.random() < 0.4 else 0 for k in range(len(line_tokens))],
            },
            "next_tokens": [next_tokens[tok] for tok in line_tokens],
        })
print('---')
print(jsonl_lines[0])
print('---')

with open("test_data.jsonl", "w") as f:
    for line in jsonl_lines:
        f.write(json.dumps(line) + "\n")
```
