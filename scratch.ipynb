```python
import random
import math
import json

# Generate some test data

vocab = ["foo", "bar", "baz", "qux", "quux", "corge", "grault", "garply", "waldo", "fred", "plugh", "xyzzy", "thud"]

jsonl_lines = []
for i in range(100):
    for j in range(random.randint(5, 10)):
        line_tokens = [random.choice(vocab) for _ in range(random.randint(5, 35))]
        jsonl_lines.append({
            "tokens": line_tokens,
            "metrics": {
                "length": [len(t) for t in line_tokens],
                "random": [random.random() * n for n, _ in enumerate(line_tokens)],
                "entropy": [random.uniform(0, 1) * math.cos(k * math.pi / 10) if random.random() < 0.4 else 0 for k in range(len(line_tokens))],
            }
        })

print(jsonl_lines)

with open("test_data.jsonl", "w") as f:
    for line in jsonl_lines:
        f.write(json.dumps(line) + "\n")
```
